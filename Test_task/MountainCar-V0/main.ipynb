```python
from train import main
import numpy as np
from eval import eval_main
```

```python
sqrt_reducer = lambda eps, i: eps/np.sqrt(i+1)
lin_reducer = lambda eps, i: eps/(i+1)
```

```python
k=0
with open('agents_description', 'w') as f:
    for reducer_name, reducer in zip(('sqrt', 'line'),(sqrt_reducer, lin_reducer)):
        for sv in (True, False):
            for rsc in (300, int(1e3), int(3e3)):
                line = '{}:, reducer {}, sv: {}, rsc: {} \n'.format('agent'+str(k), reducer_name, sv, rsc)
                f.write(line)
                main('agent'+str(k), reducer, sv, rsc)
                print('____________________________________________')
                k += 1
```

```python
eval_main(['agent'+str(i) for i in range(k)])
```
